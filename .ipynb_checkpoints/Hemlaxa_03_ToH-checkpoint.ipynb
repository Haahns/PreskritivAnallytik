```python
# Initial values
# Antal tillstnd: 3^3
V = []
for i in range(27):
    V.append(0)
V[26] = 1
```

```python
#Define all states
indexDict =[]

actions = [
    [1,2], #1 -> 2, 3
    [0,2,5], #2 -> 1, 3, 6
    [0,1,3], #3 -> 1, 2, 4
    [2,4,7], #4 -> 3, 5, 8
    [3,7,8], #5 -> 4, 8, 9
    [1,6,9], #6 -> 2, 7, 10
    [5,7,9], #7 -> 6, 8, 10
    [3,4,6], #8 -> 4, 5, 7
    [4,10,11], #9 -> 5, 11, 12
    [5,6,12], #10 -> 6, 7, 13
    [8,11,14], #11 -> 9, 12, 15
    [8,10,15], #12 -> 9, 11, 16
    [9,16,18], #13 -> 10, 17, 19
    [14,17], #14 -> 15, 18
    [10,13,17], #15 -> 11, 14, 18
    [11,20,21], #16 -> 12, 21, 22
    [12,18,19], #17 -> 13, 19, 20
    [13,14,20], #18 -> 14, 15, 21
    [12,16,24], #19 -> 13, 17, 25
    [16,22,23], #20 -> 17, 23, 24
    [15,17,21], #21 -> 16, 18, 22
    [15,20,23], #22 -> 16, 21, 24
    [19,23,25], #23 -> 20, 24, 26
    [19,21,22], #24 -> 20, 22, 23
    [18,25,26], #25 -> 19, 26, 27
    [22,24,26], #26 -> 23, 25, 27
    [24,25] # ml!
]
```

```python
# Initial policy
policy = []
for i in range(27):
    policy.append(np.random.choice(actions[i]))
```

```python
#Hyperparameters
SMALL_ENOUGH = 0.5
GAMMA = 0.9         
NOISE = 0.10
```

```python

iteration = 0

while iteration < 100:
    biggest_change = 0
    for s in range(27):
        
        old_v = V[s]
        new_v = 0
            
        for t in actions[s]:
            nxt = t
            
            #Choose a new random action to do (transition probability)
            act=np.random.choice(actions[s])
            print('t:',t,'s:',s,'nxt:',nxt,'act:',act)
            #Calculate the value
            v = V[s] + (GAMMA * ((1-NOISE)* V[nxt] + (NOISE * V[act]))) 
            if v > new_v: #Is this the best action so far? If so, keep it
                new_v = v
                policy[s] = t
        print('Old V:',old_v,'New V:',new_v)
       #Save the best of all actions for the state                                
        V[s] = new_v
        biggest_change = max(biggest_change, np.abs(old_v - V[s]))
        #print(biggest_change)
            
    #See if the loop should stop now         
    if biggest_change < SMALL_ENOUGH:
        break
    iteration += 1
print(iteration)
```

```python
%pylab inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

def show_state(n):
    s = str(n)
    if n < 10: s = '0' + s
    img = mpimg.imread('img/'+ s +'.png')
    imgplot = plt.imshow(img)
    plt.show()
```

```python
i = 0
show_state(i+1)
while True:
    # Vlj det av tillgngliga tillstnd som har hgst V
    options = []
    for a in actions[i]:
        options.append(V[a])
    i = V.index(max(options))
    show_state(i+1)
    if i == 26:
        break
```
